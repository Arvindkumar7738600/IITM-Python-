```python
##Q3:Sum of squares of even number.


#Approch -->  Evens number ---> square ---> sum

def sum_of_squares_of_evens_number(nums):
    evens = []
    for i in nums:
        if i%2 == 0:

            evens.append(i)

    for i in range(len(evens)):
        evens[i] = evens[i]**2

    return sum(evens)

nums = [1, 2, 3, 4, 5]
print(sum_of_squares_of_evens_number(nums))


# Now one more option to tackel this question in other and short way: Using of Concepts : Filter and map:

def sum_of_squares_of_even_numbers(nums):

    def is_even(n):
        return n%2 == 0

    def square(n):
        return n**2

    evens = filter(is_even, nums)
    squares = map(square, evens)

    return sum (squares)

x = [1, 2, 3, 4, 5]
print(sum_of_squares_of_even_numbers(x))
```

```python
# Q1: Write a program to merge two dictionaries if a key is present in both dictionaries, sum the values. Assume the values are numbers (int or float)


'''
d1= {'a': 1, 'b': 2, 'c': 3}
d2= {'b':3, 'c': 4, 'd': 5}
merge_dict(dl, d2) â†’ {'a': 1, 'b': 5, 'c': 7,'d': 5}
'''


d1 = {'a': 1, 'b': 2, 'c': 3}
d2 = {'b': 3, 'c': 4, 'd': 5}

#merge: {'a': 1, 'b': 5, 'c': 7, 'd': 5}

def merge_dicit(d1, d2):
    d = {}
    for k in d1:
        d[k] = d1[k]

    for k in d2:
        if k in d:
            d[k] += d2[k]
        else:
            d[k] = d2[k]

    return d


print(merge_dicit(d1, d2))
```

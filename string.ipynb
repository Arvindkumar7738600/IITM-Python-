#------Membership Operator------

What are Membership Operators?
Python provides two membership operators:
in: Returns True if a value is found in the sequence. ---- SYNTAX:  (valus in sequence)-----

not in: Returns True if a value is not found in the sequence.  -----SYNTAX:  (valus not in sequence)----


```python
fruits=["apple", "banana", "cherry"]

#check if banana is in the list 
print("banana" in fruits)
print("mango" in fruits)
```

```python
#Check if "grape" is not in the list
print("grape" not in fruits)
print("banana" not in fruits)
```

```python
#Check in Number Functions
numbers= (1,2, 3,4,5)
print(3 in numbers) #True
print(6 not in numbers) #True
```

```python
#Check in String Functions
message="Hello, world!"
print("Hello" in message)
print("world" not in message)
```

#----- Caesar Cipher Technique -------#

'''
The Caesar Cipher is one of the simplest and most widely known encryption techniques.
It is a type of substitution cipher where each letter in the plaintext is shifted by a certain number of places down or up the alphabet.
For example, with a shift of 3:

A becomes D
B becomes E
Z becomes C (wraps around the alphabet)

How the Caesar Cipher Works:

---Encryption: Each letter in the plaintext is shifted by a fixed number (key) to produce the ciphertext.

---Decryption: Each letter in the ciphertext is shifted back by the same number to retrieve the plaintext.

---% 26 ensures the shift is within the range of the alphabet (0-25).
'''

```python
alpha="abcdefghijklmnopqrstuvwxyz"
s='india'

t=''
i=0
k=2
t=t +((alpha[(alpha.index(s[i])+k)%26]))
t=t +((alpha[(alpha.index(s[i+1])+k)%26]))
t=t +((alpha[(alpha.index(s[i+2])+k)%26]))
t=t +((alpha[(alpha.index(s[i+3])+k)%26]))
t=t +((alpha[(alpha.index(s[i+4])+k)%26]))

print(t)
```

#----if-elif-else Statement----#
'''The elif (else if) statement checks multiple conditions.'''

```python
x=int(input("Enter a Number"))
if x>10:
    print(x, "is greater than 10")
elif x==10:
    print(x, "is equal to 10")
else:
    print(x, "is less than 10")
```

#----Using Logical Operator----#

```python
#and
'''Returns True if both conditions are True.'''

x=int(input ("Enter a number"))
if x>5 and x<15:
    print(x, "is between 5 and 15")
```

```python
#or
'''Returns True if at least one condition is True.'''

x=int(input ("Enter a number"))
if x>5 or x<15:
    print(x, "is outside the range 5 - 15")
```

```python
#not
'''Reverses the result of a condition.'''

x=int(input("Enter a Number"))
if not x==5:
    print(x, "is not equal to 5")
```

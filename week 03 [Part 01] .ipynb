Problem Statement:
Write a Python program that calculates the factorial of a given non-negative integer.

For example, given the number 5, the program should compute:
# 5!= 5*4*3*2*1

```python
answer=1
answer=answer*2
answer=answer*3
answer=answer*4
answer=answer*5
print("The factorial of 5 is", answer)
```

```python
number=int(input("Enter a non-negative integer: "))
i=1
answer=1
while (i<=number):
    answer=answer * i
    i=i+1
print ("The factorial of", number, "is", answer)
```

```python
#while loop

num=int(input ("Enter a number: "))
absNum=abs (num)
rev=absNum%10
absNum=absNum//10
while(absNum>0):
    r=absNum%10
    absNum=absNum//10
    rev=rev*10 +r
if(num<0) :
    rev=rev-2*rev
if(num==rev):
    print( "Palindrome" )
else:
    print("Not a palindrome")
```

Nested For Loop in Python

Definition:

A nested for loop is a loop placed inside another loop.
The outer loop iterates over its sequence, and for each iteration of the outer loop, the inner loop runs completely.
This structure is especially useful for processing multi-dimensional data or generating combinations.

```python
for i in range(3): #Outer for 1oop 
    for j in range(3): #Inner for loop
        print("i=",i, "j=",j)
```

Problem Statement

Write a Python program to find and print all prime numbers less than a number entered by the user.

A prime number is a natural number greater than 1 that has no divisors other than 1 and itself.

```python
num=int(input("Enter a number: "))

if(num>2):
    print(2, end=" ")

for i in range(3, num):
    flag=False

    for j in range (2,i):
        if(i%j == 0):
            flag=False
            break
        else:
            flag=True
    if flag:
        print(i, end=" ")
```

The break statement immediately terminates the nearest enclosing loop (for or while) regardless of the loop's condition.

After the break, the program resumes execution with the statement immediately following the loop.

continue
Definition:
The continue statement skips the current iteration of the loop and immediately returns execution to the top of the loop for the next iteration, leaving any code after continue in the loop body unexecuted for that iteration.



for item in iterable:
    if condition:
        continue # Skip the rest of this iteration and move to the next
     Other statements that run only if condition is False


Example:
Imagine you want to print only the odd numbers from a range:

```python
#1
#3
#5
#7
#9


for i in range (10):
    if i % 2 ==0:
        continue #skip even numbers
    print(i)
    
```

pass
Definition:

The pass statement is a null operation-it does nothing when executed.
It is used as a placeholder when a statement is required syntactically, but you do not want to perform any action.

This is commonly used as a placeholder for future code, in empty functions, classes, or loops.

```python
for x in range(11):
    if(x%3==0):
        print(x)
    else:
        pass
```

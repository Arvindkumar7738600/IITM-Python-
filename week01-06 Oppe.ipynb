```python
'''<------Slicing------->'''

# Concept of slicing - [start:ends:step]

name = "sweetiee"  # i want to print Sweet
print(name[0:5])

x = "substring"   # I want to print string 
print(x[3:])

y = "Parampreet"
# I want to print ram
print(y[2:5:1]) # 2, 2+1 = 3, 3+1 = 4, 4+1 = 5 STOPPING HERE ...

# I want to print ram
print(y[2:5:2]) # 2, 2+2 = 4, 4+2 = 6,  STOPPING HERE ...

# I want to print rampreet
print(y[2::1])  # Start = 2 , Ends = Length of y , Step = 1

# I want to print Param
print(y[:5:1])  # Start = 0, Ends = 5 , Step = 1

# print odd number
print(y[::2])

# print even number
print(y[1::2])

print(y[9:])  # Start = 9, Ends= ? , Step = 1


print(len(y[10:])) # Start = 10, Ends= ? , Step = 1 


print(y[10:]) # Start = 10, Ends= ? , Step = 1 
# it means this is print a empty .
```

```python
'''------Reverse slicing-------'''

#Concept : start = -1, Ends = -n , Steps = -1, -2, -3, -4, -5, -6 -n

name = "Parampreet"
print(name[::-1])   # Start = -1 , Ends = -10, Steps = -1 (-1 means is reverse order of slicing)


print(name[-1:-11:-1])   # Start = -1 , Ends = -11, Steps = -1 (-1 means is reverse order of slicing)

#-1，-14-1 =-2，-2+（-1）=-3，... , -9+(-1)=-10, -10+(-1)=-11 STOPPING HERE.. (not considering -11)


print(name[-1:-11:-2])   # Start = -1 , Ends = -11, Steps = -2 (-1 means is reverse order of slicing)


print(name[-2:-11:-2])   # Start = -2 , Ends = -11, Steps = -2 (-1 means is reverse order of slicing)

```

```python
'''------Any------'''

lst = [False, True, True, True, False]
print(any(lst)) # gives us true, if atleast single element has True value, else False
print(all(lst)) # gives us true, only when ALL THE ELEMENTS has True value, else False
```
